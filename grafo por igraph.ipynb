```python
import networkx as nx
import matplotlib.pyplot as plt
from igraph import Graph

# Carrega o grafo do arquivo .graphml
graph = Graph.Read_GraphML('/content/sao jose campestre.graphml')
#graph = Graph.Read_GraphML('/content/Meium (site).graphml')
#graph = Graph.Read_GraphML('/content/Santana do serido.graphml')

# Calcula as centralidades
eigenvector_centrality = graph.eigenvector_centrality()
betweenness_centrality = graph.betweenness()
degree_centrality = graph.degree()
closeness_centrality = graph.closeness()

# Converte para NetworkX para facilitar a visualização
nx_graph = graph.to_networkx()

# Função para plotar o grafo com uma centralidade específica
def plot_graph(centrality, title):
    plt.figure(figsize=(10, 10))
    node_size = [v * 1000 for v in centrality]  # Ajusta o tamanho dos nós baseado na centralidade
    nx.draw(nx_graph, node_size=node_size, with_labels=False, node_color=node_size, cmap=plt.cm.Blues)
    plt.title(title)
    plt.show()

# Exibe os grafos
#plot_graph(eigenvector_centrality, "Eigenvector Centrality")
#plot_graph(betweenness_centrality, "Betweenness Centrality")
#plot_graph(degree_centrality, "Degree Centrality")
plot_graph(closeness_centrality, "Closeness Centrality")
```
